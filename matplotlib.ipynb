```python
import matplotlib.pyplot as plt
```

# Pie Chart

```python
areas = ['Marketing', 'Sales', 'Development', 'HR', 'Customer Support']
budget = [2.5, 4, 10, 1, 6]
```

```python
plt.pie(budget, labels = areas, autopct = "%0.2f%%")
plt.show()
```

```python
x = [23, 45, 12, 49, 97, 32, 11, 44]
y = [45, 34, 67, 87, 55, 22, 89, 90]
plt.scatter(x, y)
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Age Vs Salary Graph')
plt.show()
```

```python
x = [1, 5, 9, 12, 22, 44, 66, 78]
y = [5, 20, 15, 6, 1, -50, 20, 30]
y1 = [51, 201, 151, 61, 11, -501, 201, 301]
y2 = [-2, -3, 500, 600, 10, 12, 1000, 2]
plt.plot(x, y, color = 'red', marker = 'D', markersize = 10, label = "one")
plt.plot(x, y1, color = 'blue', marker = '+', markersize = 10, label = "two")
plt.plot(x, y2, color = 'green', marker = 'D', markersize = 10, label = "three")
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Age Vs Salary 2d Plot')
plt.legend()
plt.show()
```

```python
x = [2012, 2013, 2014, 2015, 2016, 2017, 2018]
y = [1000, 2000, 3000, 4000, 5000, 6000, 7000]
plt.bar(x, y, color = 'red')
plt.xlabel('year')
plt.ylabel('bikes sold')
plt.title('Bikes sold per year')
plt.show()
```
