**Cumulative Frequencty (Prefix Frequency)**

```python
freq = data['Pclass'].value_counts()


sum = freq.sum() #get total sum 
#Cumulative Frequency in Persentage ( Total Is 100%)
#we divide all frequecies by total sum and multiply with 100 to get corresponding percentages
percentageFreq = ((freq/sum)*100).values
#so total percentage is 100

#cumulative frequencty in percentage 
cumFreq = []
temp = 0
for i in percentageFreq:
  temp = temp+i
  cumFreq.append(temp)
#in cumulative frequency , can get all previous values 
plt.plot(cumFreq,marker='D') #For plot an array or list , we can use plt.plot() function
#for better understand the curve we give a marker (D denote diamond shape)

```

# Measure of Spread 
Measure of Spread gives us an idea of how much the data of spread or varying . Some important measure of spread are 

1. Range
2. Interquartile Range(IQR)
3. Variance
4. Standard Deviation

```python
# Here we work with iris dataset 

#Range -> describe the difference between largest and smallest point of data

range = np.max(iris_setosa['PL']) - np.min(iris_setosa['PL']) #max - min
print("The range is ",range)

#percentile -> a measure used in statistics indicating the value below which a given percentage of observations in a group of observations falls. 
percentile = np.percentile(iris_setosa['PL'],70)
print("70 percentile of iris_setora[PL] is ",percentile) #it means 70% of data below 1.3

#Quartiles -> are the values that divide a list of number into quarters
#cut the data into x equal part , the quartiles are at the cuts
#let we cut 4 equal part
q1 = np.percentile(iris_setosa['PL'],25)
q2 = np.percentile(iris_setosa['PL'],50)
q3 = np.percentile(iris_setosa['PL'],75)
q4 = np.percentile(iris_setosa['PL'],100)
print(f"First Quartile : {q1}, second quartile : {q2}, third quarlite : {q3}, forth quartile : {q4}")


#Interquartile range -> is a measure of statistical dispersion between upper(75th) and lower(25th) quartiles

iqr = np.percentile(iris_setosa['PL'],75) - np.percentile(iris_setosa['PL'],25)
print(f"Interquartile range is {iqr}")


#Mean Absolute Deviation or absolute deviation from mean -> describe the variation in the dataset , it tells the average absolute distance of each data point in the set 

#variance -> measure how far are data points spread out from the mean
#the variance is computed by finding the differences between every point of data and the mean , squaring them , summing them up , and taking the average those numbers
#the squares are used during the calculation because the weight outliers more heavily than points are the near to the mean . This prevent the differences above the mean
#netralize those below the mean

variance=np.var(iris_setosa['PL'])
print(f"Variance is {variance}")

#but we dont use variance to find out the spread out the data , instead of variance , we use standard deviation
#problem of variance is : because of the squaring , it is not in the same unit of measurement as the original data

#for this ,we use standard deviation , it is the square root of variance 
sd=np.std(iris_setosa['PL'])
print(f"standard deviation is {sd}")


#median absolute deviation -> is the median of all the number obtained from subtracting and calulating the absolute value of every observation with the median
from statsmodels import robust
mad = robust.mad(iris_setosa['PL'])
print(f"Mad is {mad}")
#If our dataset contains outliers , we prefer mad instead of variance

#boxplot -> gives a complete overview of the dataset such as minimum value,maximum value, outliers , interquartile ranges , median etc .. 
sns.boxplot(x='Species' , y = 'PL', data = data)


```

#Probability Density Function(PDF)
If we have one numerical variable , we use histograms for visualization. But histograms have a problem. In histograms , it is difficult to  analysis multiclass numrical data , data for different classes can be overlapped. Using PDF we can easily analysis multiclass numerical data.
Probability Density Funtion provides density of the points in a particular range. It is calculated using karnel density estimations techniques . 

For calculatig PDF of a paticular point , we find the frequency of the points and divide by the total samples


```python
#distplot-> provides probability density with histogram
sns.distplot(iris_setosa['PL'])
#we can use FacetGrid to draw distplot side by side for multiclass numerical data
#here we try to analysis PL of  all three Species 
sns.FacetGrid(data, hue='Species' , height= 6).map(sns.distplot,'PL').add_legend() 
```

#Cumulative Distibution Function
For calculating cdf a paricular point x, we count number of samples less equal to the point x , then divide by the total number of samples . Another way is take prefix of pdf. Another way is using calculus . using Area under the curve techniques we can find cdf. CDF of a paritcular point x  would be Area under the pdf curve uptill that point. If we integrate pdf, we get cdf. and if we differentiate cdf, we get pdf.

```python
#histogram -> here we are not plot histogram , we take the values and store into counts and binedges
counts , bin_edges = np.histogram(iris_setosa['PL'],bins=10)

pdf = counts/sum(counts); 

cdf = np.cumsum(pdf);

plt.plot(bin_edges[1:],pdf) #plot(x,y)
plt.plot(bin_edges[1:],cdf)

#using cdf , we can  find what is the percentage of correct guesses or wrong guesses
```

```python
#how standardization is performed 
from scipy.stats import norm

#now generate 100000 normally distributed random integers with specified mean and std
normal_random = norm.ppf(np.random.random(100000),loc = 1000, scale = 200).astype(int)  #mean 1000, std = 200

#now plot this
sns.distplot(normal_random)

```

```python
temp = []
for i in normal_random:
  temp.append((i-1000)/200)
new_normal = np.array(temp)
print(np.round(new_normal.mean()))
print(np.round(new_normal.std()))
sns.distplot(new_normal)

#here we convert normally distributed numbers with mean = 1000 , std = 200 to mean = 0 and std = 1 , which is standard normal distribution/variate
```

#Central Limit Theorem  
The central limit theorem states that if you have a population with mean μ and standard deviation σ and take sufficiently large random samples from the population with replacementtext annotation indicator, then the distribution of the sample means will be approximately normally distributed. This will hold true regardless of whether the source population is normal or skewed, provided the sample size is sufficiently large (usually n > 30).
