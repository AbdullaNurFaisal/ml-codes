<center><h3><b>Scatter Plots</b></h3></center>
The scatter plot is mainly of statistical visualization . It depicts the joint distribution of two variables using a cloud of points,where each point represents an observation in the dataset. This depiction allows the eye to infer a substantial amount of information about whether there is any meaningful relationship between them. It just basically gives a high level overview of data. <br>
Although, scatter plot can be plotted for both numerical as well as catergorical data, it's mostly used to depics relationship between numerical data.


There are 2 ways of draw a scatterplot in seaboarn.

    1. relplot (Figure Level Function
    2. scatterplot (Axis Level Function)

These visualization can be quite helpful because they use simple and easily understood representations of data even on complex data by mapping up to three additional variables using the semantics of <b> hue , size , and style.</b>

```python
#libray 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
plt.style.use('fivethirtyeight') #import a style named fivethirtyeight
```

```python
data = sns.load_dataset('tips') #sns has some example dataset, tips is one of them , we use this for demonstration
```

```python
data.head() #overview of data
```

**<h3>Using the relplot function</h3>**

<b> Plot a Sctter Plot between total_bill and tip </b>

```python
#we use replot for plot scatterplot
sns.relplot(x='total_bill',y='tip', data=data)
#we can add kind param
#sns.relplot(x='total_bill',y='tip',kind='scatter', data=data)
```

**The Hue parameters**

```python
#we want to show the distribution based on the basis smoker column
#we pass a catergorical data as hue parameters
sns.relplot(x='total_bill', y = 'tip', hue='smoker', data=data)
```

<b> The style paramaters </b>


```python
#more information 
sns.relplot(x='total_bill', y = 'tip', hue='smoker',style='sex', data=data)
```

<b>The size parameters</b>

```python
#can pass numerical data
#more information
sns.relplot(x='total_bill', y = 'tip', hue='smoker',style='sex',size='size', data=data)
```

```python
#can customize sizes
sns.relplot(x='total_bill', y = 'tip', hue='smoker',style='sex',size='size',sizes=(15,200), data=data)
```

**<h3>Using the scatterplot function</h3>**

```python
sns.scatterplot(x='total_bill',y='tip', data=data)
```

```python
sns.scatterplot(x='total_bill', y = 'tip', hue='smoker', data=data)
```

<b>Boxplots</b>

A boxplot is a standardized way of displaying the distribution of data based on a five number summary (“minimum”, first quartile (Q1), median, third quartile (Q3), and “maximum”). It can tell you about your outliers and what their values are. It can also tell you if your data is symmetrical, how tightly your data is grouped, and if and how your data is skewed.

**<b>Violin Plot</b>**

Violin plots are a method of plotting numeric data and can be considered a combination of the boxplot with a kernel density plot.
The advantage of the violin plot is , it gives us the power of boxplot and kde(PDE). 
Violin plot are often used to compare the distribution of a given variable across some categories


```python
sns.violinplot(tips['total_bill'])
```

```python
sns.catplot(x='day',y='total_bill',kind='violin',data=tips)
```

```python
sns.catplot(x='day',y='total_bill',kind='violin',hue='sex',data=tips)
```

```python

sns.catplot(x='day',y='total_bill',kind='violin',hue='sex',split=True,data=tips)
```

**<center><h3>Catergorical estimate plots</h3></center>**

<b>Barplot and Countplot</b>

These graph rather than giving information about the distribution of data, they give data about central tendency.

```python
#mean value of total_bill  of smoker and non smoker
sns.catplot(x='smoker',y='total_bill',kind='bar',data=tips)
```

```python
sns.catplot(x='smoker',y='total_bill',hue='sex',kind='bar',data=tips)
```

```python
#for median, we change estimator value , by default it is set for mean , it can also change to variance or std
sns.catplot(x='smoker',y='total_bill',hue='sex',estimator=np.median,kind='bar',data=tips)
```

Counplot is special case of barplot ,where do not consider numerical data.
Takes categorical values and find counts of these values. 
It is just like a histogram for categorical data

```python
#number of males and females 
sns.catplot(x='sex',kind='count',data=tips)
```

```python
#how many male/female are smoker and non smoker
sns.catplot(x='sex',kind='count',hue='smoker',data=tips)
```

<b>PointPlot</b>
<br>
A point plot represents an estimate of central tendency for a numeric variable by the position of scatter plot points and provides some indication of the uncertainty around that estimate using error bars.
point plot shows only the mean (or other estimator) value.

Ref: https://seaborn.pydata.org/generated/seaborn.pointplot.html


```python
sns.pointplot(x="time", y="total_bill", data=tips)
```

```python
ax = sns.pointplot(x="time", y="total_bill",hue='sex', data=tips)
```

**<center><h3> Categorical Categorical Plots </h3></center>**



<b>Heatmaps</b>
<br>
A heatmap is a graphical representation of data where the individual values contained in a matrx are represented as colors.

```python
#here we use flight dataset
flights=sns.load_dataset('flights')
flights.head()
```

```python
ax = flights.pivot_table(index='year',columns='month',values='passengers',aggfunc='sum')
ax
```

```python
sns.heatmap(ax)
#sns.heatmap(ax, cbar=False)
```

```python
sns.heatmap(ax, cbar=False,linewidths=0.5,annot=True, fmt='d')
```

```python
#can change color tone
sns.heatmap(ax, cbar=False,linewidths=0.5,annot=True, fmt='d',cmap='winter')
```

<b>Clustermaps</b><br>
It is basically a heatmap, the only difference is that, it is clustered.
It clustered row and column together on the basis of similarity.

```python
sns.clustermap(ax)
#by defult it uses matric as euclidean , we can change it
```

```python
#we can normalize the score using z_score parameters
sns.clustermap(ax,z_score=0,annot=True)
```

**<center><h3>JointPlot</h3></center>**
<br>
Jointplit creates a multi-panel figure that shows both the bivariate (or joint) relationship between two variables along with the univariate (or marginal) distribution of each on separate axes.

We can use jointplots 4 ways,
  1. ScatterPlot
  2. Hexbin plots
  3. Kdeplots
  4. Regplot


```python
#use tips dataset
sns.jointplot(x='total_bill',y='tip',data=tips) #it is a scatterplot with both axis histogram
```

```python
sns.jointplot(x='total_bill',y='tip',kind='hex', data=tips) #it is a 2d histogram
```

```python
sns.jointplot(x='total_bill',y='tip',kind='kde', data=tips) #2d kde plot
```

```python
sns.jointplot(x='total_bill',y='tip',kind='reg', data=tips) #regration plot
```

**<center><h3>Pairplot</h3></center>**
<br>
Pairplot is used to plot multiple pairwise bivariate distributions in a dataset. This creates a matrix of axes and shows the relationship for each pair of columns in a DataFrame. By default, it also draws the univariate distribution of each variable on the diagonal Axes.

```python
#use iris dataset
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue='species') #here we can figure out what columns are strongly related and what column is important 
```

**<center><h3>Distplots</h3></center>**
<br>
Distplots are used to do univariate analysis on a continious/numerical data.
Distplot combines 3 things in one graph:
1. Histogram
2. Kernel Density Estimation(PDF)
3. Rugplot

```python
#using titanic dataset
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic['age'].fillna(titanic['age'].mean(),inplace=True) #this dataset has some mission values in age . we replace missing ages with mean 
```

<b>Plotting a Distplot</b>

```python
sns.distplot(titanic['age'])
```

**Plotting only the Histogram**

```python
sns.distplot(titanic['age'],kde=False)
```

**Plotting only the KDE**

```python
sns.distplot(titanic['age'],hist=False)
```

**Plotting only the Rugplot**

```python
sns.distplot(titanic['age'],kde=False,hist=False,rug=True)
```

**How to use the Distplot**

```python
sns.distplot(titanic[titanic['survived']==1]['age']) #fetch all value of ages that are survived
```

```python
#side by side comparison
sns.distplot(titanic[titanic['survived']==0]['age'])
sns.distplot(titanic[titanic['survived']==1]['age'])
```
